```python
# Librerías

import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# Carga de predicciones

```python
predicciones = pd.read_csv('model_predictions.csv')
```

```python
predicciones.head()
```

```python
# Extraigo las etiquetas verdaderas y las predicciones de los modelos
y_true = predicciones['true_labels']
y_pred_nb = predicciones['nb_predictions']
y_pred_lr = predicciones['lr_predictions']
```

# Reporte de métricas

```python
def report_metrics_and_confusion_matrix(y_true, y_pred, model_name):
    print(f"Resultados del modelo {model_name}:")
    print(classification_report(y_true, y_pred))
    # Matriz de confusión
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Negativo', 'Positivo'], yticklabels=['Negativo', 'Positivo'])
    plt.xlabel('Predicción')
    plt.ylabel('Real')
    plt.title(f'Matriz de Confusión - {model_name}')
    plt.show()

report_metrics_and_confusion_matrix(y_true, y_pred_nb, "Naive Bayes")

report_metrics_and_confusion_matrix(y_true, y_pred_lr, "Regresión Logística")
```

# Conclusiones finales

- Exactitud (Accuracy):
Ambos modelos tienen una precisión similar, con Naive Bayes obteniendo un 83% y Regresión Logística un 84%. Esto indica que ambos modelos son capaces de clasificar correctamente la mayoría de las muestras.

- Precisión, Recall y F1-Score:
La precisión, recall y F1-Score son ligeramente mejores en el modelo de Regresión Logística para ambas clases.

Ambos modelos muestran un desempeño equilibrado, lo cual sugiere que no hay un sesgo significativo en la clasificación de ambas clases.

- Matriz de confusión:
La matriz de confusión para ambos modelos muestra un buen equilibrio entre verdaderos positivos y verdaderos negativos, confirmando que ambos modelos son efectivos en distinguir entre clases positivas y negativas. Siendo el equilibrio mayor en la Regresión Logística.

- Selección del modelo:
Dado que la Regresión Logística muestra una ligera ventaja en términos de precisión, recall y F1-Score, se recomienda seleccionar este modelo para el análisis de sentimiento de estas reviews de Amazon.

- Optimización:
Aunque los resultados son buenos, hay margen para la optimización. Esto podría incluir la búsqueda de hiperparámetros, el uso de técnicas de reducción de dimensionalidad o la incorporación de características adicionales.
